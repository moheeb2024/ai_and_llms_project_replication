# For the manifestos from 1998 to 2025, there are 48 CSV files. We combine them all here in one csv file. 

```python
import os
print(os.getcwd()) # to ensure we're in the right directory
```

```python
# ChatGPT was used to assist in combining 48 CSV files into a single dataset and to add the metadata to the file. The prompt used was: 
# “I have 48 CSV files in one folder. How can I combine them into one file? Note: they all have the same number and names of columns.” 

import pandas as pd
import glob
import os

csv_files = [f for f in glob.glob("*.csv") ]

df_list = []

for file in csv_files:
    df = pd.read_csv(file)

    filename = os.path.basename(file).replace(".csv", "")
    parts = filename.split("_")

    if len(parts) < 2:
        print(f"Skipping unexpected filename: {filename}")
        continue

    party_id = parts[0]
    date_code = parts[1]
    year = date_code[:4]

    df["party_id"] = party_id
    df["year"] = year

    df_list.append(df)

combined_df = pd.concat(df_list, ignore_index=True)

combined_df["sentence_id"] = (
    combined_df["party_id"].astype(str) + "_" +
    combined_df["year"].astype(str) + "_" +
    combined_df.index.astype(str)
)

combined_df.to_csv("combined.csv", index=False)
```

```python
# We have here a problem with the cmp_code variable, which is the variable of interest for the analysis. The cmp_code variable is currently in a numeric format, but it should be in a string format. This is because the cmp_code variable contains codes that are not purely numeric, and we want to preserve the leading zeros in the codes. Therefore, we need to clean the cmp_code variable and convert it to a string format. 
# To do this, we asked ChatGPT: “How can I clean the cmp_code variable in my dataset to convert it to a string format and remove any leading zeros?” and we provided a pic of how this variable looks like. The response was:


combined_df["cmp_code_clean"] = combined_df["cmp_code"].astype(str).str.strip() # to create a new variable called cmp_code_clean, which is a cleaned version of the cmp_code variable. This is done by converting the cmp_code variable to a string and stripping any leading or trailing whitespace.

combined_df["cmp_code_clean"] = combined_df["cmp_code_clean"].str.replace(".0", "", regex=False) # to remove the ".0" from the cmp_code_clean variable, which is a common issue when reading numeric codes from CSV files. This is done by replacing ".0" with an empty string.
```

```python

# We received also this prompt from ChatGPT to use gpt-5-mini for classification. This code was used as a base and we  changed the text of the prompt and the number of testing rows to see how it works.
# Prompt 1

import json
import time
import pandas as pd
from openai import OpenAI


df = pd.read_csv("combined.csv") # to read the combined dataset. 
TEXT_COL = "text_en" # this is the name of the variable that contains the text that we want to classify as migration-related or non-migration-related.

for col in ["gpt_migration", "gpt_confidence"]: # to add two new variables to the dataset, one for the classification result (gpt_migration) and one for the confidence score (gpt_confidence). These variables will be filled with the results from gpt-5-mini.
    if col not in df.columns:
        df[col] = pd.NA

client = OpenAI()

def build_prompt(text): # the prompt text 
    return f"""
Task:

You are coding political text for a binary classification task.
Task:
Determine whether the text is MIGRATION-RELATED or NON-MIGRATION_RELATED
Code 1 = migration-related 
Code 0 = non-migration-related 
Use only the text provided. Do not infer any hidden context.
Classify a text as MIGRATION-RELATED (1) only if it explicitly refers to at least one of the following:
-	immigration, emigration, migration 
-	migrants, immigrants, emigrants 
-	asylum, asylum seekers 
-	refugees 
-	deportation, expulsion, removal 
-	borders, border control, border protection in relation to the movement of people 
-	residence permits, visas, naturalisation, citizenship for immigrants 
-	integration of migrants or refugees 
-	family reunification of migrants or refugees 
-	irregular/illegal migration, undocumented migrants 
-	immigration quotas or the admission of foreigners 
Classify a text as NON-MIGRATION_RELATED (0) if the text is mainly about:
-	general welfare, education, health, housing, labour, economy 
-	crime or security in general 
-	foreign policy in general 
-	the EU in general 
-	development aid 
-	human rights in general 
-	national identity or culture without explicit reference to migrants/refugees/asylum
-	mobility/transport unrelated to migration 
-	tourism or travel 
Important decision rules:
-	Mention of foreigners, refugees, asylum seekers, immigrants, or border control for human movement => 1
-	General references to “integration” without mention of migrants, refugees, or immigrants => 0
-	General references to “security” or “borders” without mentioning migrants/refugees/immigrants => 0
-	General references to “citizenship” or “nationality” without migrant context => 0
-	Only choose 1 when migration is explicit and central, not merely implied

Return JSON only.

Text:
{text}
""".strip()

def classify_row(text, model="gpt-5-mini"): # the function to classify a single row of text using gpt-5-mini. The function takes the text and the model name as input and returns the classification label and confidence score as output.
    response = client.responses.create(
        model=model,
        input=build_prompt(text),
        text={
            "format": {
                "type": "json_schema",
                "name": "migration_classifier",
                "schema": {
                    "type": "object",
                    "properties": {
                        "label": {
                            "type": "integer",
                            "enum": [0, 1]
                        },
                        "confidence": {
                            "type": "number",
                            "minimum": 0,
                            "maximum": 1
                        }
                    },
                    "required": ["label", "confidence"],
                    "additionalProperties": False
                },
                "strict": True
            }
        }
    )

    result = json.loads(response.output_text)
    return result["label"], result["confidence"]


start_row = 0 # starting from first row. 
end_row = min(100, len(df)) # to set the end row. 
save_every = 100

for i in range(start_row, end_row): # to loop through the rows of the dataset and classify each row using the classify_row function. The results are stored in the gpt_migration and gpt_confidence variables in the dataset. The loop also includes error handling to retry classification up to 3 times in case of failures, with a delay between attempts.
    if pd.notna(df.at[i, "gpt_migration"]):
        continue

    text = df.at[i, TEXT_COL]
    if pd.isna(text) or not str(text).strip():
        continue

    for attempt in range(3):
        try:
            label, confidence = classify_row(str(text))
            df.at[i, "gpt_migration"] = label
            df.at[i, "gpt_confidence"] = confidence
            print(f"Row {i}: OK")
            break

        except Exception as e:
            print(f"Row {i}: attempt {attempt + 1} failed -> {e}")
            time.sleep(2) # wait before retrying

df.iloc[:100].to_excel("prompt_1_first_100.xlsx", index=False) # to save the first 100 rows of the dataset with the new classification results in an Excel file. This allows us to check the results and see if the classification is working as expected before applying it to the entire dataset.
print("Final file saved: prompt_1_first_100.xlsx") # to print a message indicating that the final file has been saved. This is useful for tracking the progress of the code execution and confirming that the results have been saved successfully.
```

```python

# We have asked ChatGPT: “How can I evaluate the performance of the classification model using the results from the prompts_1_7 dataset? I want to compare the predicted labels with the manually coded labels and calculate metrics such as accuracy, precision, recall, and F1-score.” The response was:
# To evaluate the performance of the classification model using the results from the prompts_1_7 dataset, you can follow these steps:

from sklearn.metrics import precision_score, recall_score, f1_score, accuracy_score # to import the necessary metrics for evaluating the performance of the classification model. 
```

```python
prompt_cols = [f"prompt_{i}" for i in range(1, 8)] # to create a list of the column names for the predicted labels from the different prompts. This will allow us to loop through these columns and calculate the performance metrics for each prompt.

results = []


for col in prompt_cols:
    y_true = prompts["checking"] # to get the true labels from the "checking" column in the prompts_1_7 dataset. This column contains the manually coded labels that we will use as the ground truth for evaluating the model's performance.
    y_pred = prompts[col]
    
    precision = precision_score(y_true, y_pred) # to calculate the precision score for the predicted labels for each prompt column compared to the true labels. 
    recall = recall_score(y_true, y_pred) # to calculate the recall score for the predicted labels for each prompt column compared to the true labels.
    f1 = f1_score(y_true, y_pred) # to calculate the F1-score for the predicted labels for each prompt column compared to the true labels. 
    accuracy = accuracy_score(y_true, y_pred) # to calculate the accuracy score for the predicted labels in each prompt column compared to the true labels.
    results.append({
        "prompt": col,
        "precision": precision,
        "recall": recall,
        "f1_score": f1,
        "accuracy": accuracy
    }) # to append the calculated performance metrics for the current prompt column to the results list as a dictionary. Each dictionary contains the prompt name and the corresponding precision, recall, F1-score, and accuracy values.
```

```python
results_df = pd.DataFrame(results) # to convert the results list, which contains dictionaries of performance metrics for each prompt, into a pandas DataFrame. 
results_df = results_df.sort_values(by="f1_score", ascending=False) # to sort the results DataFrame by the F1-score in descending order. 
print(results_df) # print the results. 
```

```python
# we wanted to check the confusion matrix for each prompt since the performance metrics can be influenced by the distribution of true positives, true negatives, false positives, and false negatives. 
# so we asked ChatGPT: “How can we check the confusion matrix for each prompt? 
from sklearn.metrics import confusion_matrix

y_true = prompts["checking"]
prompt_cols = [f"prompt_{i}" for i in range(1, 8)]

for col in prompt_cols:
    y_pred = prompts[col]
    cm = confusion_matrix(y_true, y_pred, labels=[0, 1])

    tn, fp, fn, tp = cm.ravel()

    print(f"\n{col}")
    print(f"TN = {tn}, FP = {fp}, FN = {fn}, TP = {tp}")
```
